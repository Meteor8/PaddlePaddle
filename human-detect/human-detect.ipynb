```python
import paddlehub as hub
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
%matplotlib inline
```

```python
humanseg = hub.Module(name="deeplabv3p_xception65_humanseg")
```

```python
path = ["data/0.jpg"]
# 注意要可视图，则需设置visualization
results = humanseg.segmentation(paths=path,visualization=True,output_dir="data")
result = results[0]

img0 = mpimg.imread(path[0])
test_img_path = result["save_path"]
img1 = mpimg.imread(test_img_path)
plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(img0)
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img1)
plt.axis("off")
plt.show()
```
