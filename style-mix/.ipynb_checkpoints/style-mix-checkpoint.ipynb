```python
import paddlehub as hub
import cv2
import matplotlib.pyplot as plt 
import matplotlib.image as mpimg 
%matplotlib inline
```

```python
stylepro_artistic = hub.Module(name="stylepro_artistic")
```

```python
results = stylepro_artistic.style_transfer(images=[{
        'content': cv2.imread(r"data/1.jpg"),
        'styles': [cv2.imread(r"data/0.jpg")]}],
        alpha = 1.0,
        visualization = True)
```

```python
result_path = results[0]['save_path']
img0 = mpimg.imread(r"data/0.jpg")
img1 = mpimg.imread(r"data/1.jpg")
img2 = mpimg.imread(result_path)

plt.subplot(1,3,1)
plt.imshow(img0)
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(img1)
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(img2)
plt.axis("off")
plt.show()
```
